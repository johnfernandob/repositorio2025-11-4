```python
def es_primo(n):
    """Devuelve True si n es un n√∫mero primo, False en caso contrario."""
    if not isinstance(n, int) or n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    limite = int(n**0.5) + 1
    for i in range(3, limite, 2):
        if n % i == 0:
            return False
    return True
es_primo(29)
```
